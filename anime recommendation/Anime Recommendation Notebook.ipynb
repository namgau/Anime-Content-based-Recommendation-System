# 3.Content-based filtering with a neural network

```python
#scale training data
animes_train_unscaled = animes_train
users_train_unscaled = users_train
y_train_unscaled    = y_train

scalerAnime = StandardScaler()
scalerAnime.fit(animes_train)
animes_train = scalerAnime.transform(animes_train)

scalerUser = StandardScaler()
scalerUser.fit(users_train)
users_train = scalerUser.transform(users_train)

scalerTarget = MinMaxScaler((-1, 1))
scalerTarget.fit(np.array(y_train).reshape(-1, 1))
y_train = scalerTarget.transform(np.array(y_train).reshape(-1, 1))
#ynorm_test = scalerTarget.transform(y_test.reshape(-1, 1))

print(np.allclose(animes_train_unscaled, scalerAnime.inverse_transform(animes_train)))
print(np.allclose(users_train_unscaled, scalerUser.inverse_transform(users_train)))
```

### Split the data

```python
animes_train, animes_test = train_test_split(animes_train, train_size=0.80, shuffle=True, random_state=1)
users_train, users_test = train_test_split(users_train, train_size=0.80, shuffle=True, random_state=1)
y_train, y_test       = train_test_split(y_train,    train_size=0.80, shuffle=True, random_state=1)
print(f"animes training data shape: {animes_train.shape}")
print(f"animes test data shape: {animes_test.shape}")
```

# 4.Neural Network

```python
num_outputs = 48

users_NN = Sequential([
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(num_outputs, activation='linear')
])

animes_NN = Sequential([
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(num_outputs, activation='linear')
])

# user input
input_users = Input(shape=(users_train.shape[1], ))  #needs tuple
vu = users_NN(input_users)
vu = Lambda(lambda x: tf.linalg.l2_normalize(x, axis=1))(vu)          #idk why we have to do this

# anime input
input_animes = Input(shape=(animes_train.shape[1], ))
va = animes_NN(input_animes)
va = Lambda(lambda x: tf.linalg.l2_normalize(x, axis=1))(va)

# dot product
output = Dot(axes=1)([vu, va])

# model
model = Model([input_users, input_animes],  output)
model.summary()
```

### mean square error loss and Adam optimizer

```python
cost_fn = tf.keras.losses.MeanSquaredError()
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt, loss=cost_fn)
```

```python
print(users_train.shape)
```

```python
print(animes_train.shape)
```

```python
tf.random.set_seed(1)
model.fit([users_train, animes_train], y_train, epochs = 50)
```

```python
model.evaluate([users_test, animes_test], y_test)
```

### Finding similar items

```python
def sq_dist (a, b):
    return sum(np.square(a - b))
```

#### compute matrix of square distance between a anime and other animes

```python
# make use of numpy masked array to avoid selecting the same movie. The masked values along the diagonal won't be included in the computation.
```

```python
vas
```

```python
count = 50
dim = len(vas)
dist = np.zeros((dim,dim))
```

```python
for i in range(dim):
    for j in range(dim):
        dist[i,j] = sq_dist(vas[i, :], vas[j, :])

m_dist = np.ma.masked_array(dist, mask=np.identity(dist.shape[0]))

m_dist
```

```python

def get_anime_genre(item, ivs, genres):
    offset = np.where(item[ivs:] == 1)[0][0]
    genre = genres[offset]
    return(genre, offset)
```

```python
disp = [["anime1", "genres", "anime2", "genres"]]
for i in range(count):
    min_idx = np.argmin(m_dist[i])
    anime1_id = (int(anime_list_bi.iloc[i]['uid']))        #idk, this should be a string
    anime2_id = (int(anime_list_bi.iloc[min_idx]['uid']))
    genre1, _ = get_anime_genre(anime_list_bi.iloc[i], 4, genres)      # 4 is where the first genre starts
    genre2, _ = get_anime_genre(anime_list_bi.iloc[min_idx], 4, genres)

    disp.append( [anime_list_genre.loc[anime_list_genre['uid'] == anime1_id, 'title'].values[0], genre1,
                  anime_list_genre.loc[anime_list_genre['uid'] == anime2_id, 'title'].values[0], genre2]
               )
table = tabulate.tabulate(disp, tablefmt='html', headers="firstrow", floatfmt=[".1f", ".1f", ".0f", ".2f", ".2f"])
table
```
